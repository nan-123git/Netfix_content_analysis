<b> Handling Missing Values
Filled missing values in important columns like director, cast, and country to maintain consistency.</b>

```python
#df['Director'].fillna('Unknown' , inplace = True) #changed in python 3.0 , it will always behave as a copy
df.fillna({'director':'Unknown', 'cast':'Not Available' , 'country' :'Unknown', 'release_Date':'Unknown' , 'rating':'Unknown'}, inplace=True)
df
```

<b> Data Transformation
Converted date column and extracted year for trend analysis.</b>

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')  #if you want 2020-08-14 format then do this if some date can't convert it will show NaT
#df
```

```python
df['year'] =df['Release_Date'].dt.year
df['year'] = df['year'].fillna(0).astype(int)  # i use fillna so that if there are any NaT values ; they don't affect the conversion
print(df['year'])
```

<b>Feature Engineering
Created new columns such as duration_num and duration_type for better analysis. </b>

```python
df['duration_num'] = df['Duration'].str.extract('(\d+)').astype(int)  #here it only takes numbers like 90,4 etc
df['duration_type'] = df['Duration'].str.extract('([a-zA-Z]+)')  #here it takes strings like seasons , min etc
```

```python
df
```
